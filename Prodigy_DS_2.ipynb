```python
import pandas as pd

titanic_data = pd.read_csv('/content/gender_submission.csv')
print(titanic_data)
```

```python
titanic_data['Survived'].fillna(titanic_data['Survived'].mean(), inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(titanic_data['Survived'], bins=30, kde=True)
plt.title('Survival count')
plt.xlabel('survived')
plt.ylabel('count')
plt.show()
```
